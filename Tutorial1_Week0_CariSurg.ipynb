```python
import pandas as pan
import numpy as py
import matplotlib.pyplot as plt

FILE_PATH = '/content/drive/MyDrive/Colab Notebooks/EmergencyTriageDataset_Reduced_Dirty.csv'

df = pan.read_csv(FILE_PATH)
df.head(11)
```

```python
print(df.dtypes)
```

```python
#Standardizing the elements to lowercase to map fewer elements
df['Gender'] = df['Gender'].astype(str)
df['Gender'] = df['Gender'].str.strip().str.lower()
print(f"\nValues in .lower() Gender Column: {df['Gender'].unique()} Total unique values: {df['Gender'].nunique()}")

#Setting up the map for every variant of Gender
genderMap = {
    'male': 1,
    '1': 1,
    'female':0,
    '0': 0
}

#Replace the intial gender column with the cleaned column
df['Gender'] = df['Gender'].map(genderMap)
print(f"\nCleaned Gender Column {df['Gender'].value_counts()}")

Nan = df['Gender'].isnull().sum()
if(Nan>0):
    print("Nan values detected!")

else:
  print("\nNo NaN detected")

#Print to see results
df.head()
```
